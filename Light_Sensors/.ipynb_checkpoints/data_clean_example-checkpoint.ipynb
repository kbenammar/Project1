```python
import pandas as pd
import numpy as np
import datetime
import json 
```

```python
file_path = "../content/test.txt"
```

```python
column_names = ["Date","Mean_light","min_light","max_light","variance"]

df = pd.read_csv(file_path, names=column_names)
# df.head()
```

```python
df.head()
```

```python
date_string_format = "%d/%B/%m|%H:%M:%S"

def split_and_take(element_guy):

  return element_guy.split(":")

for ind,row in df.iterrows():
  # case for datetime
  #split
  date_parts = row["Date"].split("':") 
  #take date part that we want
  date = date_parts[1]
  # print(date)
  # print(type(date))
  #turn into datetime object
  date_ = date.replace("|"," ")
  # date_object = datetime.datetime.strptime(date,date_string_format)
  #replace
  row["Date"] = date_

  # case where no '{' or '}'
  # split and take last element in new array 
  row['Mean_light'] = row['Mean_light'].split(":")[1]
  row['min_light'] = row['min_light'].split(":")[1]
  row['max_light'] = row['max_light'].split(":")[1]

  #case for '}'
  #split
  variance_parts = row['variance'].split(":")
  #strip
  variance = variance_parts[1].strip("}")
  #replace
  row["variance"] = variance
  
  # row["Mean_light"] = float(split_and_take(row['Mean_light']))
  # row['min_light'] = float(split_and_take(row['min_light']))
  # row["max_light"] = float(split_and_take(row["max_light"]))
  # row["variance"] = float(split_and_take(row['variance']))
  
  
```
