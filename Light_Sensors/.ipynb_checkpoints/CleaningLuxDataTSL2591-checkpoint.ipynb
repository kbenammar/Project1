```python
# importing libraries :
import pandas as pd
import numpy as np
import datetime
import json 
```

### Cleaning the data ###

```python
# setting file path
filepath="luxmeasuresTSL2591"
```

```python
# making a list of the column names
column_names=["date","mean_lux","min_lux","max_lux","variance_lux","mean_visible","min_visible","max_visible","variance_visible"]
```

```python
# loadind file into DataFrame
df=pd.read_csv(filepath, names=column_names)
```

```python
df
```

```python
df.info()
```

```python
df.tail()
```

```python
for ind,row in df.iterrows():

    # case for datetime
    #split
    date_parts = row["date"].split("':") 
    #take date part that we want
    date = date_parts[1]
    #turn into datetime format object
    date_ = date.replace("|"," ")
    #replace
    row["date"] = date_
    
    # case where no '{' or '}'
    # split and take last element in new array 
    row['mean_lux'] = row['mean_lux'].split(":")[1]
    row['min_lux'] = row['min_lux'].split(":")[1]
    row['max_lux'] = row['max_lux'].split(":")[1]
    row['variance_lux'] = row['variance_lux'].split(":")[1]
    row['mean_visible'] = row['mean_visible'].split(":")[1]
    row['min_visible'] = row['min_visible'].split(":")[1]
    row['max_visible'] = row['max_visible'].split(":")[1]
    
    #case for '}'
    #split
    variance_parts = row['variance_visible'].split(":")
    #strip
    variance = variance_parts[1].strip("}")
    #replace
    row["variance_visible"] = variance
```

```python
df
```

```python
# converting date and time to datetime type
df['date'] = pd.to_datetime(df["date"])
```

```python
df['date']= df['date'].apply(lambda t: t.strftime("%y-%m-%d %H:%M:%S"))
```
