```python
# importing libraries :
import pandas as pd
import numpy as np
import datetime
import json 
```

```python
# importing files to merge:
filepath1="cleanData_TSL2591.txt"
filepath2="cleanData_VEML7700.txt"
```

```python
# creating dataframes:
df_left=pd.read_csv(filepath1)
df_right=pd.read_csv(filepath2)
```

### Cleaning left dataframe before merge ###

```python
df_left
```

```python
# convert date to datetime dtype
df_left['date'] = pd.to_datetime(df_left["date"])
```

```python
# to apply specific datetime dtype function
df_left['date_min']=df_left['date'].apply(lambda t: t.strftime("%y-%m-%d %H:%M"))
df_left.head()
```

```python
# changing name of columns before merge
df_left['mean_lux_tsl'] = df_left['mean_lux']
df_left['min_lux_tsl'] = df_left['min_lux']
df_left['max_lux_tsl'] = df_left['max_lux']
df_left['variance_lux_tsl'] = df_left['variance_lux']
```

```python
# subsetting only the wanted columns
df_left=df_left[['date_min','mean_lux_tsl','min_lux_tsl','max_lux_tsl','variance_lux_tsl','mean_visible','min_visible','max_visible','variance_visible']]
```

```python
df_left.head()
```

### Cleaning right dataframe before merge ###

```python
df_right
```

```python
# convert date to datetime dtype
df_right['date'] = pd.to_datetime(df_right["date"])
```

```python
# to apply specific datetime dtype function
df_right['date_min']=df_right['date'].apply(lambda t: t.strftime("%y-%m-%d %H:%M"))
df_right.head()
```

```python
# changing name of columns before merge
df_right['mean_lux_veml'] = df_right['mean_lux']
df_right['min_lux_veml'] = df_right['min_lux']
df_right['max_lux_veml'] = df_right['max_lux']
df_right['variance_lux_veml'] = df_right['variance_lux']
```

```python
# subsetting only the wanted columns
df_right=df_right[['date_min','mean_lux_veml','min_lux_veml','max_lux_veml','variance_lux_veml','mean_light','min_light','max_light','variance_light']]
```

```python
df_right.head()
```

### Meging dataframes on date_min key column ###

```python
new_df=pd.merge(df_left, df_right, on= "date_min", how= 'inner')
new_df
```

### creating a function to plot a column as such

```python
def fplotLux(dataFrame,column):
    return dataFrame.plot(x='date_min',y=str(column),figsize=(12,9),title= 'lux level during 24h',kind='line',xlabel='date(YY-MM-DD hh:mm)', ylabel='lux',rot=45,stacked=False)
```

```python
fplotLux(df_left,'mean_lux_tsl')
```
