```python
conn=sqlite3.connect('SQLData_TSL.sqlite')
cursor=conn.cursor()
print("Opened Database successfully")
```

```python
cursor.execute("""CREATE TABLE TSL_MEASURES
                (ID INT PRIMARY KEY NOT NULL,
                DATE DATETYPE NOT NULL,
                MEAN_LUX_TSL FLOAT NOT NULL,
                MIN_LUX_TSL FLOAT NOT NULL,
                MAX_LUX_TSL FLOAT NOT NULL,
                VARIANCE_LUX_TSL FLOAT NOT NULL,
                MEAN_VISIBLE FLOAT NOT NULL,
                MIN_VISIBLE FLOAT NOT NULL,
                MAX_VISIBLE FLOAT NOT NULL,
                VARIANCE_VISIBLE FLOAT NOT NULL);""")
```

```python
ID=0
for i,row in df.iterrows():
    ID+=1
    cursor.execute("INSERT INTO TSL_MEASURES (ID,DATE,MEAN_LUX_TSL,MIN_LUX_TSL,MAX_LUX_TSL,VARIANCE_LUX_TSL,MEAN_VISIBLE,MIN_VISIBLE,MAX_VISIBLE,VARIANCE_VISIBLE) values(?,?,?,?,?,?,?,?,?,?)",(ID,row['date'],row['mean_lux'],row['min_lux'],row['max_lux'],row['variance_lux'],row['mean_visible'],row['min_visible'],row['max_visible'],row['variance_visible']))

    
```
